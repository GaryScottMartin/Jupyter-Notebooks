#### MARGIN OF ERROR (ME)

One of the main caveats of sample results is that they vary from sample to sample, so you shouldn’t report statistical results without also including the margin of error. 

The __margin of error__ is the amount by which you expect the sample results to vary from sample to sample. A large margin of error means you expect the sample results to change a lot (not a good thing), and a small margin of error means you don’t expect the sample results to change much with repeated sampling (a good thing).
In this course, you tackle problems involving the margin of error for the sample mean or for the sample proportion. 

__Quantitative Data:__ If you want to determine the average value for the population, you use the sample mean, plus or minus the margin of error for your sample mean.

__Categorical Data (yes/no, or success/failure):__ If you want to determine the percentage in the population with a certain characteristic, you use the sample proportion, plus or minus the margin of error for your sample proportion.

The margin of error formulas look different on the surface in each situation, but they each contain three important components: the $Z^*$ value, The standard deviation of the population, and the sample size. 

Note: the $Z^*$ value is a number from the standard normal distribution and reflects the number of standard errors you need to add/ subtract in order to get the confidence level you want. For example, to be 95 percent $(95\%)$ confident, you need to add/subtract about $ 2.0 \cdot $ standard error ($1.96$ to be exact) because of the Empirical Rule.

The specific formula for **__margin of error__ for the __sample mean__** is: $ \pm \left( Z_{\alpha / 2} \right) ^* \cdot \left( \frac{\sigma}{\sqrt{n}} \right) $

The specific formula for **__margin of error__ for the __sample proportion__** is: $ \pm \left( Z_{\alpha/2} \right) ^* \cdot \left( \sqrt{\frac{\hat{p}(1-\hat{p})}{n}} \right) $

##### INTERPRETING MARGIN OF ERROR CORRECTLY

Margin of error measures the level of precision in your sample results. Precision represents consistency, but it doesn’t necessarily promise on-target results (they may be systematically over or under by a certain amount each time). In other word, margin of error doesn’t measure the amount of bias that may come with the sample results. Most types of bias are extremely difficult to measure, so you should make an effort minimize bias wherever possible.

>

##### __Should We planning on using Cis for Two population means for …..?__

Suppose a group of 100 people is on one weight loss plan for one month and then switches to another weight loss plan for one month. Suppose you want to compare the two plans, and you decide to compare the average weight loss for the first month to the second month.
* Should we  plan on using a 95% confidence interval for the average difference or not? Explain.

> *__I think the design of the experiment is flawed and that a comparison based on Confidence Intervals could be misleading. In this particular experimental design there are likely to be confounding variables. The primary one being that most strategies for weight loss cause the body to adapt by reducing the basal metabolic rate as time passes. Therefore, it is very likely that the second plan would unfairly be penalized just because the subjects have been on another plan for a month before starting and reduced basal metabolic rate would reduce the results of that second plan. A better design would be to divide the group into two and have each half try an assigned plan simultaneously, that would at least eliminate the bias resulting from reduced basal metabolic rate. At that point a two mean CI estimate might be useful in comparing the results of the two plans.__*
