#### Random Sampling Techniques

1. Simple Random Sample (SRS) := every item in the population has an equal probability of selection.  

2. Systematic Sampling (SyS):= Simple Random Sampling in an ordered systematic way.  

3. Stratified Sampling (StrS): Population divided into groups from which we sample randomly. (Provides information on subgroups)  

4. Cluster Sampling (CS): Population is divided into (geographical) cluster. Some specific clusters are chosen by SRS. Individual samples within cluster units are chosen via SRS. Clusters are assumed homogeneous.  

###### Non-random Sampling Techniques
1. Self-Selected (Voluntary) Sampling := Inclusion or exclusion of samplling units determined by units themselves, either implictly or explicitly.
2. Convenience Sampling := A sample of items which are easy to reach.

> Another way to summarize categorical data is to show the percentage of
individuals that fall into each category, thereby creating a relative frequency.

sample size, multiplied by 100 to get the percentage. If you list all the possible
categories along with their relative frequencies, you create a relative frequency
table. The total of all the relative frequencies should equal 100 percent (subject
to possible round-off error)  
> Relative frequencies do just what they say: They help you relate the results to

#### WHAT ARE THE MEASURES OF CENTER?

There are several different ways to determine the center, so we have different definitions of measures of center, including the mean, median, mode, and midrange.

-  The mean (arithmetic mean) of a set of values is the measure of center found by adding the values and dividing the total by the number of values.
> <p>$x$ is the variable usually used to represent the individual data values.</p>
$ \newcommand{\olsi}[1]{\,\overline{\!{#1}}} % overline short italic $  
> <p>$\olsi{x} = \frac{\sum_{i=1}^n x_i}{n}$ is the mean of a set of sample value. n is the number of values in a sample</p>
-  The median of a data is the measure of center that is the middle value when the original data values are arranged in order of increasing (or decreasing) magnitude. 
> *  If the number of values is odd, the median is the value in the exact middle. (i.e., the $\frac{n+1}{2}th$ ordered observations)
> *  If the number of values is even, the median is found by computing the mean of the two middle numbers. (i.e., the average of  $\frac{n}{2}th$ and $(\frac{n}{2}+1)th$ ordered observations.

- The mode of a data set, often denoted by M, is the value that occurs most frequently.
  - When two values occur with the same greatest frequency, each one is a mode and the data set is bimodal.  
  
-  The midrange is the measure of center that is the value midway between the highest and lowest values in the original dataset.  

#### WHAT ARE THE DISTRIBUTIONS OF DATA?

Data skewed to the left have a longer left tail, and the mean and median are to the left of the mode. Although not always predictable, data skewed to the left generally have a mean less than the median.

A distribution of data is symmetric if the left half of its histogram is roughly a mirror image of its right half.

Data skewed to the right have a longer right tail, and the mean and median are to the right of the mode. Although not always predictable, data skewed to the right generally have a mean to the right of the median.

#### WHAT ARE THE MEASURES OF VARIATION?

_There are several different ways to determine the center, so we have different definitions of measures of variation, including the __range__, __standard deviation__, __variance__, and __coefficient of variation__._
<br/>
<ul>
    <li>The <b>range</b> of a set of data is the difference between the highest value and the lowest value.</li><br/>
    <li>The <b>standard deviation</b> of a set of sample values is a measure of variation of values about the mean. It is a type of average deviation of values from the mean that is calculated by using<br/>
$\sigma=\sqrt\frac{\sum_{i=1}^n(x_i-\olsi{x})^2}{n-1}$ is the sample standard deviation</li>
</ul>

<u>How to find the standard deviation?</u>
<ol>
<li>Compute the mean $\olsi{x}$</li>
<li>Subtract the mean from each individual value to get a list of deviations $(x_i-\olsi{x})$.</li>
<li>Square each of the difference obtained from step 2. This produces numbers of the form $(x_i-\olsi{x})^2$</li>
<li>Add all of the squares obtained from Step 3. This is the value of $\sum_{i=1}^n(x_i-\olsi{x})^2$</li>
<li>Divide the total from Step 4 by the number $(n-1)$</li>
<li>Find the square root of the result of Step 5</li>
</ol>
<br/>
<ul>
    <li>The <b>variance</b> of a set of values is a measure of variation equal to the square of the standard deviation.<br/>
Sample variance $(\sigma^2)$: Square of the sample standard deviation $\sigma$</li>
<br/>
    <li>The <b>coefficient of variation</b> (<b>CV</b>) for a set of sample or population data, expressed as a percent, describes the standard deviation relative to the mean, and is given by the following:<br/>  
            Sample: $CV=\frac{\sigma}{\olsi{x}} \cdot 100\%$</li>
</ul>

#### WHAT IS THE EMPIRICAL (OR 68-95-99.7) RULE?  



For data sets having a distribution that is approximately bell-shaped (data sets that are approximately normally distributed), the following properties apply  

- About 68% of all values fall within 1 standard deviation of the mean  
- About 95% of all values fall within 2 standard deviation of the mean  
- About 99.7% of all values fall within 3 standard deviation of the mean

#### Probability

##### Definitions

A __probability__ is the underlying foundation on which the important methods of inferential statistics are built.

* An __event__ is any collection of results or outcomes of a procedure.
* The __sample space__ for a procedure consists of all possible events.
* __Notation for Probabilities__  
$P$ denotes a probability.  
$A$, $B$, and $C$ denote specific events.  
$P(A)$ denotes the probability of event $A$ occurring.  

* __Relative Frequency Approximation of Probability__  
Conduct (or observe) a procedure a large number of times and count the number of times that event A actually occurs. Based on these actual results, P(A) is estimated as follows:  
> $P(A)=\frac{\text{number of occurrences of}\,A}{\text{number of trials}}$  
* __Classical Approach to Probability (Requires Equally Likely Outcomes)__    
Assume that a given procedure has n different simple events and that each of those simple events has an equal chance of occurring. If event A can occur in s of these ways, then  
> $P(A)=\frac{\text{number of ways}\,A\,\text{can occur}}{\text{number of unique simple events}}=\frac{s}{n}$  
* __Subjective Probabilities__  
$P(A)$, the probability of event $A$, is found by simply guessing or estimating its value based on knowledge of the relevant circumstances.
* __Law of Large Numbers__  
As a procedure is repeated again and again, the relative frequency probability of an event tends to approach the actual probability.

##### Probability Rules
* Probability Rules:
  * __The probability of an impossible event is 0.__  
  * __The probability of an event that is certain to occur is 1.__  
  * $0<=P(A)<=1$ for any event A  
  * $P(\lnot A)=1-P(A)$  
  * $P(A \lor B) = P(A) + P(B) - P(A \land B)$
  * $A$ and $B$ are two mutually exclusive events if $P(A \land B)=0$   
  * $A$ and $B$ are two independent events if $P(A \land B)=P(A) \cdot P(B)$ (or $P(A|B)=P(A)$ or $P(B|A)=P(B)$ )
  
#### Random Variables & Probability Distribution

##### Random Variables

A *__random variable__* is a variable (typically represented by x) that has a single numerical value, determined by chance, for each outcome of a procedure.  
-	*__Discrete random variable__*: A discrete random variable has either a finite number of values or a countable number of values, where “countable” refers to the fact that there might be infinitely many values, but they can be associated with a counting process.  
-	*__Continuous random variable__*: A continuous random variable has infinitely many values, and those values can be associated with measurements on a continuous scale without gaps or interruptions.  

##### Probability Distribution
A *__probability distribution__* is a description that gives the *__probability__* for each value of the *__random variable__*. It is often expressed in the format of a graph, table, or formula.

###### Requirements for a Probability Distribution
1. The sum of all probabilities must be 1
2. For every individual value of x, the corresponding probability must be between 0 and 1 inclusive.

###### Mean and Standard Deviation for probability distribution

1. *__Mean (Expected Value)__* for a probability distribution. The mean of a discrete random variable is the theoretical mean outcome for infinitely many trials. <u>*__We can think of that mean as the expected value in the sense that is the average value that we would expect to get if the trials could continue indefinitely.__*</u>  

> $\mu=\sum[x \cdot P(x)]$  

2.	__Variance__ for a probability distribution  

> $\sigma^2=\sum[(x-\mu)^2 \cdot P(x)]$

3.	__Standard Deviation__ for a probability distribution  

> $\sigma=\sqrt{\sum[(x-\mu)^2 \cdot P(x)]}$

## MISSION 05 &mdash; BINOMIAL DISTRIBUTION (DISCRETE PROBABILITY DISTRIBUTION)

### SUMMARY OF MISSION 5

#### Conditions required for binomial Distribution

1. Fixed number of trials, denoted by n.
2. Trials must be independent.
3. Two possible outcomes for each trial $(\text{success} \lor \text{failure})$
4. $p$ (which denotes the probability of success) remains constant from trial to trial.  

#### Notation for Binomial Distribution and Find Binomial Probability 

Notation for binomial distribution:  

| | |  
| :- | :- |  
|$n$|The number of times a trial is repeated|  
|$p=P(S)$|The probability of success in a single trial|  
|$q=P(F)=1-p$|The probability of failure in a single trial|  
|$x$|a count of the number of successes in n trials|  


#### Methods of Evaluating Probabilities of Binomially DIstributed Data Sets

##### Binomial Table  

The probability of $x$ successes in any series of $n$ Bernoulli Trials where $x, n,$ and $p$ are known and where $n$ is relatively small may be looked up in a binomial probability table. A portion of such a table is shown here.  



##### Enumerate the Outcomes, Compute the Probabilities of Each, Consolidate into Probability Distribution Function Table  

There are exactly $2^n$ possible outcomes of any particular series of Bernoulli Trials. Where $n$ is relatively small, it is reasonable to enumerate each possible outcome, directly compute its probability, and then consolidate the probabilities of the outcomes into a probability distribution function table. This table may then be used to lookup the probability of a particular number of successes, $x$, out of $n$ trials or to compute the probability of any particular set of outcomes (e.g. $a \le x \le b$). An example of this method is shown here.  



##### Binomial Formula  

Binomial Formula: In a binomial experiment, the probability of exactly $x$ successes in $n$ trials is:  

> $P(x)=C_x^n p^x q^{n-x}$ where:  

| | |  
| :- | :- |  
|$C_x^n=\frac{n!}{x!(n-x)!}$|The number of ways of getting x successes in n trials|  
|$n$|The number of times a trial is repeated|  
|$p=P(S)$|The probability of success in a single trial|  
|$q=P(F)$|The probability of failure in a single trial|  
|$x$|a count of the number of successes in n trials|  

Predefined implementations of the Binomial Formula are available for use in many calculators or software applications and languages such as Excel, Google Sheets, or Python.

#### Expected Value, Variance, Standard Deviation, and shape of Binomial Distribution

| | |  
| :- | :- |  
| Expected Value (mean) | $ \mu = E(X) = n \cdot p $ |  
| Variance | $ \sigma^2 = Var(X) = n \cdot p \cdot q = n \cdot p \cdot (1-p) $ |  
| Standard Deviation | $ \sigma = Sd(X) = \sqrt{n \cdot p \cdot q} = \sqrt{n \cdot p \cdot (1-p)} $ |  
|Symmetric | $ p=0.5 $|  
|Skewed Right|$ p<0.5 $|  
|Skewed Left|$ p>0.5 $|  
| | |  
| Maximum Usual Value | $ \mu + 2 \sigma $ |  
| Minimum Usual Value | $ \mu - 2 \sigma $ |  
| Probability | $ P(x) = \frac{n!}{x!(n-x)!} \cdot p^x q^{n-x}, for x=0...n $ |  
| | $n$: number of trials |  
| | $x$: number of successes out of n trials |  
| | $p$: probability of success in any single trial |  
| | $q$: probability of failure in any single trial $=(1-p)$ |  
| | Minimum & Maximum Usual Value valid if and only if $ n \cdot p \ge 5 $ and $ n \cdot q \ge 5 $ |

```python
# Python code (using math, numpy, & pandas libraries)
# Computing Expected Value and Standard Deviation
import math
import numpy as np
from scipy.stats import binom
import pandas as pd
import matplotlib.pyplot as plt

# Set parameters: \# trials = 6, probability of success = 0.40 
n, p = 16, 0.20

# Compute moments
mu, sigma_sq, skew = binom.stats(n, p, moments='mvs')

# Initialize all lists/arrays as empty
c=[]
x = []
p_of_x = []
x_p_of_x =[]

# define two dimensional array
# with binomial probability mass function (PMF)
for k in range(n+1):
    c.append("x=" + str(k))
    x = np.append(x,k)
    temp=binom.pmf(k, n, p, loc=0)
    p_of_x=np.append(p_of_x,temp)
    x_p_of_x=np.append(x_p_of_x,k*temp)

d1 = np.append(x,np.append(p_of_x,x_p_of_x))
d2 = d1.reshape(3, n+1)

# Load data into pandas DataFrame
df = pd.DataFrame(data = d2,
                  index = ["x", "P(x)", "x*P(x)"],  
                  columns = c) 
print("n="+str(n))
print("p="+str(p))
print("\r\n")
print("a. Find the mean and standard deviation for the number of correct answers for such students.")
print()
print("\tExpected Value (mean) ="+str(mu))
print("\tStandard Deviation ="+str(math.sqrt(sigma_sq)))
print("\r\n")
print("b. Would it be unusual for a student getting 10 correct answers?")
print()
print("\tMinimum usual value = "+str(mu - 2*math.sqrt(sigma_sq)))
print("\tMaximum usual value = "+str(mu + 2*math.sqrt(sigma_sq)))
print("\r\n")
print("c. What is the probability that a student has 16 correct answers?")
print()
print("\tP(x)=16)="+str(df.at["P(x)", "x=16"]))
print("\r\n")
print("d. What is the probability that a student has 14 or more correct answers?")
print()
print("\tP(x)>=14)="+str(df.at["P(x)", "x=14"]+df.at["P(x)", "x=15"]+df.at["P(x)", "x=16"]))
print("\r\n")
print("e.  If Mary claimed that all her answers were guesses and she received 14 correct answers. What would you conclude?")
print()
print("\tP(x)=14)="+str(df.at["P(x)", "x=14"]))

```

#### Z-Scores  

The Z-score denotes the distance of a particular value $x$ from the mean ($\mu$) of the distribution of random variable $X$ in multiples of the standard deviation ($\sigma$) of the distribution of $X$.  
> $\text{Z-score} = Z = \frac{x-\mu}{\sigma}$

## MISSION 07 -- TRANSFORMATION  

### SUMMARY OF MISSION 07  

#### Z-Scores  


The Z-score denotes the distance of a particular value $x$ from the mean ($\mu$) of the distribution of random variable $X$ in multiples of the standard deviation ($\sigma$) of the distribution of $X$.  

#### Transformation from X to Z - Find Z-score given a value from random variable

- $z=\frac{x-\mu}{\sigma}$  

#### Transformation from Z to X - Find value for a random variable given a specific Z-score  

- $x=\mu + (z \cdot \sigma)$
